## 1.1 Background

Sebuah hotel di Portugal mengalami penurunan pendapatan tahunan sebesar ~37% akibat pembatalan reservasi. Setiap kamar yang tidak terisi adalah perishable inventory berpotensi pendapatan hilang selamanya. Manajemen tidak memiliki sistem peringatan dini untuk mengidentifikasi pemesanan berisiko tinggi. 

Akibatnya:

- Tidak bisa melakukan overbooking yang terukur.

- Tidak bisa menawarkan intervensi preventif (misal: upsell, reminder, deposit non‑refundable).

- Efisiensi operasional terganggu (staffing, persediaan, revenue management).

[*Industri perhotelan*](https://link.springer.com/article/10.1007/s40558-025-00349-9) merupakan sektor jasa yang sangat bergantung pada kemampuan dalam mengelola permintaan pelanggan secara efektif.Permintaan terhadap kamar hotel tidak bersifat konstan, melainkan berfluktuasi seiring waktu akibat berbagai faktor seperti musim, perilaku wisatawan, serta dinamika ekonomi. Oleh karena itu, pemahaman terhadap pola permintaan menjadi aspek penting dalam menjaga kinerja operasional dan finansial hotel.

Salah satu mekanisme utama dalam pengelolaan permintaan adalah sistem reservasi, di mana pelanggan melakukan pemesanan sebelum tanggal kedatangan. Meskipun sistem ini membantu perencanaan, fleksibilitas yang diberikan kepada pelanggan juga memunculkan potensi pembatalan reservasi. [**Tingginya tingkat pembatalan dapat menyebabkan ketidaksesuaian antara permintaan yang diperkirakan dengan realisasi hunian**](https://www.hftp.org/news/4128776/free-hotel-cancellations-smart-strategy-or-revenue-risk)

---

## 1.2 Business Objective

Berdasarkan pemahaman bisnis yang telah diuraikan, project ini bertujuan untuk mengembangkan pendekatan berbasis data dalam mengelola risiko pembatalan reservasi hotel. 

Tujuan : 

Membangun model prediktif yang dapat mengidentifikasi reservasi dengan probabilitas pembatalan >50%, sehingga hotel dapat melakukan intervensi tepat waktu dan mengurangi revenue loss akibat pembatalan minimal 20% dalam 6 bulan.

Secara spesifik, **Use Case Operasional yang ingin dicapai adalah:**

1. Early Warning System : Setiap reservasi baru diberi skor risiko.

Kami menindaklanjuti reservasi risiko tinggi dengan pertimbangan awal akan:

- Menawarkan non‑refundable rate dengan diskon kecil.

- Mengirim pengingat otomatis H‑7, H‑3, H‑1.

- Memasukkan kamar ke program overbooking jika risiko >70%.

2. Dynamic Overbooking : Hanya reservasi dengan probabilitas cancellation >0.8 yang di‑overbook (batas aman). Kompensasi tamu yang kena walk disiapkan dalam biaya operasional.

## 1.4 Key Performance Indicators (KPI) yang Terpengaruh Pembatalan


Pembatalan reservasi memiliki dampak langsung terhadap kinerja operasional dan finansial hotel. Untuk mengevaluasi dampak tersebut, digunakan beberapa indikator kinerja utama [**Key Performance Indicators/KPI**](https://hospitalityinsights.ehl.edu/evaluating-hotel-performance) yang berkaitan dengan tingkat hunian, pendapatan, serta efisiensi operasional.

Berikut adalah KPI utama yang relevan beserta dampak pembatalan terhadap masing-masing metrik:

| Metrik | Rumus | Dampak Pembatalan |
|--------|-------|-------------------|
| **Cancellation Rate** | (Jumlah Booking Dibatalkan / Total Booking) × 100% | Meningkat secara langsung; pada kasus ini mencapai sekitar 37%, menunjukkan tingginya ketidakpastian permintaan |
| **Occupancy Rate** | (Jumlah Kamar Terisi / Total Kamar Tersedia) × 100% | Menurun akibat kamar yang tidak terisi setelah pembatalan |
| **Average Daily Rate (ADR)** | Total Room Revenue / Total Kamar Terjual | Dapat tertekan karena hotel harus menurunkan harga untuk mengisi kamar yang kosong |
| **Revenue per Available Room (RevPAR)** | Total Room Revenue / Total Kamar Tersedia | Menurun akibat kombinasi penurunan okupansi dan potensi penyesuaian harga |
| **Gross Operating Profit per Available Room (GOPPAR)** | (Total Revenue − Operating Costs) / Total Kamar Tersedia | Menurun karena pendapatan berkurang sementara sebagian biaya tetap |

Secara keseluruhan, peningkatan *cancellation rate* akan memicu efek berantai terhadap penurunan *occupancy rate*, yang pada akhirnya berdampak pada penurunan *ADR*, *RevPAR*, serta profitabilitas hotel secara keseluruhan.

Oleh karena itu, kemampuan untuk memprediksi pembatalan reservasi secara akurat menjadi kunci dalam menjaga stabilitas kinerja KPI dan mendukung pengambilan keputusan yang lebih proaktif.

## 2.1 Permasalahan Utama (Primary Problem) Hypothesis
Tingginya tingkat pembatalan reservasi **(cancellation rate)** menyebabkan ketidakpastian dalam perencanaan okupansi, yang berdampak langsung terhadap potensi kehilangan pendapatan **(revenue loss)** dan inefisiensi operasional hotel.

Untuk mengatasi permasalahan tersebut, diperlukan pendekatan berbasis data yang mampu mengidentifikasi reservasi dengan risiko pembatalan sejak awal proses pemesanan.

Oleh karena itu **Permasalahan utama dalam project ini adalah:**
<blockquote style="background: #CDDDF6; padding: 10px; border-left: 4px solid #A6C5F2; margin: 10px 0;">

    <li>sehingga hotel dapat melakukan intervensi proaktif (seperti reminder, penyesuaian kebijakan deposit, atau strategi overbooking) guna meminimalkan revenue loss.</li>

Model yang dikembangkan bertujuan sebagai early warning system yang dapat membantu tim sales dan operasional dalam mengambil keputusan berbasis risiko.

# 4. Evaluation Metrics  

></div>
****

Konteks: Klasifikasi biner

Target: `is_canceled` = 1 (batal), 0 (tidak batal)

Tujuan utama model: Mengidentifikasi reservasi berisiko tinggi sebelum pembatalan terjadi, sehingga hotel dapat melakukan intervensi (reminder, verifikasi deposit, overbooking).

<div style="padding:6px 10px; margin:12px 0; background-color:#368412; color:white; font-weight:600; border-radius:6px;">
Metric Evaluation
</div>

<p>
Evaluasi model difokuskan pada <b>optimasi keputusan overbooking</b>, di mana hotel perlu menyeimbangkan antara memaksimalkan revenue dan menjaga reputasi.
</p>

<p>
Masalah ini merupakan <b>klasifikasi biner</b>:
</p>

<ul>
<li><b>1</b> = dibatalkan</li>
<li><b>0</b> = tidak dibatalkan</li>
</ul>

<hr>

<b>1. Precision —> Mengontrol Risiko Overbooking (Prioritas Utama)</b>

<ul>
<li><b>Precision = TP / (TP + FP)</b></li>
<li>Mengukur seberapa akurat prediksi pembatalan</li>
</ul>

<p><b>Konsekuensi False Positive (FP):</b></p>
<ul>
<li>Model memprediksi batal, tetapi tamu tetap datang</li>
<li>Hotel melakukan overbooking -> terjadi double booking</li>
<li>Dampak: <b>biaya kompensasi + penurunan reputasi</b></li>
</ul>

<p><b>Insight:</b></p>
<ul>
<li>Kesalahan FP memiliki <b>cost ~3x lebih besar</b> dibanding FN</li>
<li>Oleh karena itu, model harus <b>menekan overbooking agresif</b></li>
<li>Digunakan <b>F-beta score (β = 0.5)</b> untuk memberi bobot 2x lebih besar pada precision dari pada recall)</li>
</ul>

<hr>

<b>2. Recall : Menghindari Kamar Kosong</b>

<ul>
<li><b>Recall = TP / (TP + FN)</b></li>
<li>Mengukur kemampuan mendeteksi booking yang benar-benar akan batal</li>
</ul>

<p><b>Konsekuensi False Negative (FN):</b></p>
<ul>
<li>Booking diprediksi aman, tetapi ternyata batal</li>
<li>Hotel tidak melakukan overbooking</li>
<li>Dampak: <b>kamar kosong -> kehilangan revenue</b></li>
</ul>

<p><b>Insight:</b></p>
<ul>
<li>Recall tetap penting untuk <b>mengurangi lost revenue</b></li>
<li>Namun, risikonya masih lebih kecil dibanding kerugian reputasi akibat overbooking</li>
</ul>

<hr>

<b>3. F-beta Score (β = 0.5)</b>

<ul>
<li>F-beta memberikan bobot lebih besar pada precision dibanding recall</li>
<li><b>β = 0.5 -> bobot precision 2x lebih besar dari pada recall</b></li>
</ul>

<p><b>Insight:</b></p>
<ul>
<li>Model dirancang untuk <b>lebih konservatif dalam memprediksi cancel</b></li>
<li>Menghindari keputusan overbooking yang berisiko tinggi</li>
</ul>

<hr>

<b>4. Confusion Matrix</b>

<ul>
<li>Menunjukkan distribusi TP, FP, TN, FN</li>
<li>Digunakan untuk memahami trade-off keputusan model</li>
</ul>

<hr>

<b>Metric Priority (Business-Oriented)</b>

<table border="1" cellpadding="5">
<tr><th>Priority</th><th>Metric</th><th>Target</th><th>Business Reason</th></tr>
<tr><td>1</td><td><b>Precision</b></td><td>≥ 0.75</td><td>Menjaga reputasi hotel dan menghindari biaya overbooking</td></tr>
<tr><td>2</td><td>Recall</td><td>≥ 0.80</td><td>Mengurangi kamar kosong dan kehilangan revenue</td></tr>
<tr><td>3</td><td>F-beta (0.5)</td><td>Optimized</td><td>Menyesuaikan trade-off sesuai cost bisnis</td></tr>
<tr><td>4</td><td>AUC-ROC</td><td>≥ 0.90</td><td>Menentukan threshold optimal untuk decision making</td></tr>
</table>

<hr>

<b>Final Insight:</b>

<ul>
<li>Hotel menerapkan strategi <b>controlled overbooking</b> berdasarkan prediksi model</li>
<li>Fokus utama adalah <b>meminimalkan risiko reputasi (precision)</b> sambil tetap menjaga occupancy (recall)</li>
<li>Model tidak hanya dinilai dari akurasi, tetapi dari <b>dampak bisnis terhadap revenue dan customer experience</b></li>
</ul>

```python
# 1. Data Cleaning 
# handling ADR outlier: cap di p99 agar tidak bias analisis
adr_cap = df['adr'].quantile(0.99)
df['adr_capped'] = df['adr'].clip(upper=adr_cap)
print(f"ADR capped at p99: {adr_cap:.1f} (original max: {df['adr'].max():.1f})")

# Flag ADR = 0 (kemungkinan data error)
df['is_zero_adr'] = (df['adr'] == 0).astype(int)
print(f"Booking dengan ADR=0: {df['is_zero_adr'].sum()} ({df['is_zero_adr'].mean()*100:.1f}%)")

# Flag booking tanpa tamu (anomali)
df['no_guest_flag'] = ((df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)).astype(int)
print(f"Booking tanpa tamu  : {df['no_guest_flag'].sum()} ({df['no_guest_flag'].mean()*100:.2f}%)")
```

```python
# 2. Feature: Non-Refund Flag 
# is_nonrefund sebagai binary feature (bukan dasar interpretasi utama)
df['is_nonrefund'] = (df['deposit_type'] == 'Non Refund').astype(int)
print(f"\nBooking Non-Refund  : {df['is_nonrefund'].sum()} ({df['is_nonrefund'].mean()*100:.1f}%)")
```

```python
# 3. Agregasi tamu & durasi 
df['total_guest'] = df['adults'] + df['children'] + df['babies']
df['stay_duration'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
```

```python
# is_family: digunakan untuk EDA, tidak masuk model (redundant dengan children)
df['is_family'] = (df['children'] + df['babies'] > 0).astype(int)
```

```python
# 4. Lead time group 
df['lead_time_group'] = pd.cut(
    df['lead_time'],
    bins=[-1, 7, 30, 90, 365, float('inf')],
    labels=['Last Minute', 'Short Planning', 'Planned', 'Advanced Planning', 'Very Early']
)
df['lead_time_group'] = df['lead_time_group'].astype('category')
```

```python
# 5. Arrival date & day type
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d', errors='coerce'
)
df['day_type'] = df['arrival_date'].dt.weekday.map(lambda x: 'Weekday' if x < 5 else 'Weekend')
```

```python
# Tabel outlier berdasarkan IQR method
# Membandingkan adr (raw) vs adr_capped untuk menunjukkan efek winsorizing
kolom_numerikal = df.select_dtypes(include=['int64','float64']).columns
kolom_numerikal = [c for c in kolom_numerikal if c not in
                   ['is_canceled','is_repeated_guest','is_family',
                    'is_nonrefund','is_zero_adr','no_guest_flag']]

rows = []
for kolom in kolom_numerikal:
    Q1 = df[kolom].quantile(0.25)
    Q2 = df[kolom].quantile(0.50)
    Q3 = df[kolom].quantile(0.75)
    batas_atas  = Q3 + 1.5 * (Q3 - Q1)
    batas_bawah = Q1 - 1.5 * (Q3 - Q1)
    n_out = ((df[kolom] > batas_atas) | (df[kolom] < batas_bawah)).sum()
    rows.append({
        'Kolom': kolom,
        'Median': round(Q2, 2),
        'Batas Bawah': round(batas_bawah, 2),
        'Batas Atas': round(batas_atas, 2),
        'Jumlah Outlier': n_out,
        'Persentase (%)': round(n_out / len(df) * 100, 2)
    })

tabel_outlier = pd.DataFrame(rows).set_index('Kolom')
display(tabel_outlier.sort_values('Jumlah Outlier', ascending=False))
```

## 6.2 Bivariate Analysis (Target: is_canceled)

></div>
****


<p>Analisis ini bertujuan untuk mengidentifikasi faktor utama yang mempengaruhi pembatalan reservasi.</p>

<hr>

<div style="padding:6px 10px; margin:12px 0; background-color:#368412; color:white; font-weight:600; border-radius:6px;">
    Konsep Perhitungan Cancellation Rate
</div>

<p>
Cancellation rate merupakan proporsi booking yang dibatalkan terhadap total booking dalam suatu kategori.
</p>

<ul>
<li><b>Definisi:</b><br>
Cancellation Rate = Jumlah Booking Cancel / Total Booking</li>

<li><b>Implementasi dalam data:</b><br>
Variabel <b>is_canceled</b> menggunakan nilai:
<ul>
<li>1 -> booking dibatalkan</li>
<li>0 -> booking tidak dibatalkan</li>
</ul>
Sehingga rata-rata (<i>mean</i>) dari kolom ini merepresentasikan persentase pembatalan</li>

<li><b>Misal:</b><br>
[1, 0, 1, 0, 0] -> mean = 0.40 -> 40% booking dibatalkan</li>

<li><b>Interpretasi:</b><br>
Nilai 0.40 berarti 40% booking dalam kategori tersebut dibatalkan, sehingga menunjukkan tingkat risiko pembatalan</li>
</ul>

<hr>

<p>
Cancellation rate digunakan sebagai <b>indikator risiko</b>, bukan jumlah, untuk memahami kategori mana yang memiliki probabilitas pembatalan lebih tinggi.
</p>

```python
def plot_cancel_rate(col, xlabel='', custom_labels=None):
    # Ensure is_canceled is numeric
    temp_df = df.copy()
    temp_df['is_canceled'] = pd.to_numeric(temp_df['is_canceled'], errors='coerce')
    temp_df = temp_df.dropna(subset=['is_canceled'])
    
    # Handle single vs multiple columns
    if isinstance(col, list):
        group_cols = col
        col_name = ' + '.join(col)
    else:
        group_cols = [col]
        col_name = col
    
    # For each group col, ensure string type for grouping
    for gc in group_cols:
        temp_df[gc] = temp_df[gc].astype(str)
    
    # Hitung cancel rate
    cancel_rate = (
        temp_df.groupby(group_cols)['is_canceled']
        .mean()
        .sort_values(ascending=False) * 100
    )
    
    plot_data = cancel_rate.reset_index()
    plot_data.rename(columns={'is_canceled': 'cancel_pct'}, inplace=True)
    
    # Gabung label jika multi kolom
    if len(group_cols) > 1:
        plot_data['combined'] = plot_data[group_cols].astype(str).agg(' | '.join, axis=1)
        x_col = 'combined'
    else:
        x_col = col_name
    
    # Custom label mapping
    if custom_labels and len(group_cols) == 1:
        plot_data[x_col] = plot_data[x_col].map(custom_labels).fillna(plot_data[x_col])
    
    # Global rate
    global_rate = temp_df['is_canceled'].mean() * 100
    
    # Plot
    plt.figure(figsize=(7,4))
    ax = sns.barplot(
        data=plot_data,
        x=x_col,
        y='cancel_pct',
        order=plot_data[x_col],
        palette='Reds_r'
    )
    
    ax.axhline(global_rate, color='navy', linestyle='--', linewidth=1.2,
               label=f'Global Avg: {global_rate:.1f}%')
    
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.text(
                p.get_x() + p.get_width()/2,
                height + 0.5,
                f'{height:.1f}%',
                ha='center',
                fontsize=9,
                fontweight='bold'
            )
    
    if xlabel:
        plt.xlabel(xlabel)
    else:
        plt.xlabel(col_name.replace('_',' ').title())
    
    note = ""
    if col == 'deposit_type':
        note = "Note: Non Refund → anomaly (very high cancellation)"
    elif col == 'lead_time_group':
        note = "Note: Longer lead time → higher uncertainty"
    
    if note:
        plt.figtext(0.5, 0.01, note, ha='center', fontsize=8, style='italic')
    
    plt.ylabel('Cancel Rate (%)')
    plt.title(f'Cancellation Rate by {col_name.replace("_"," ").title()}', fontweight='bold')
    plt.xticks(rotation=25, ha='right')
    plt.legend(loc='upper right')
    sns.despine()
    plt.tight_layout(rect=[0,0.03,1,0.95])
    plt.show()
```

```python
# ── Seasonality: Cancel Rate per Bulan
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
df['arrival_date_month_cat'] = pd.Categorical(
    df['arrival_date_month'], categories=month_order, ordered=True
)

temp = df.copy()
temp['is_canceled'] = pd.to_numeric(temp['is_canceled'], errors='coerce')
monthly = temp.groupby('arrival_date_month_cat')['is_canceled'].mean().reset_index()
monthly.columns = ['Month', 'cancel_pct']
monthly['cancel_pct'] *= 100
global_avg = temp['is_canceled'].mean() * 100

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Cancel rate per bulan
ax = axes[0]
bars = ax.bar(monthly['Month'], monthly['cancel_pct'],
              color=['#d73027' if v > global_avg else '#4575b4' for v in monthly['cancel_pct']],
              edgecolor='black', linewidth=0.6)
ax.axhline(global_avg, color='navy', linestyle='--', linewidth=1.2,
           label=f'Global Avg: {global_avg:.1f}%')
ax.set_title('Cancel Rate per Bulan Kedatangan', fontweight='bold')
ax.set_xlabel('')
ax.set_ylabel('Cancel Rate (%)')
ax.tick_params(axis='x', rotation=45)
ax.legend()
for bar, val in zip(bars, monthly['cancel_pct']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.4,
            f'{val:.1f}%', ha='center', fontsize=7.5, fontweight='bold')

# Volume booking per bulan
volume = temp.groupby('arrival_date_month_cat').size().reset_index(name='count')
axes[1].bar(volume['arrival_date_month_cat'], volume['count'],
            color='#74add1', edgecolor='black', linewidth=0.6)
axes[1].set_title('Volume Booking per Bulan', fontweight='bold')
axes[1].set_ylabel('Jumlah Booking')
axes[1].tick_params(axis='x', rotation=45)

sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Pilih kolom numerik untuk correlation matrix
# Exclude: kolom EDA saja, flag binary yang bukan ML feature, dan kolom leakage
exclude_corr = [
    'arrival_date_year',          # year informatif tapi sering noise
    'arrival_date_day_of_month',  # granularity terlalu tinggi
]
num_cols = [c for c in df.select_dtypes(include=['int64','float64']).columns
            if c not in exclude_corr]
print(f"Kolom untuk correlation matrix ({len(num_cols)}):")
print(num_cols)
```
